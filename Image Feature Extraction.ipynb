```python
import cv2, os
import matplotlib.pyplot as plt
imagePath = 'Image/beach.jpg'

#read
image = cv2.imread(imagePath)

#Write
cv2.imwrite('Image/beach.png',image)

#Display/Show
plt.imshow(image)
```

```python
#The Default color mode in OpenCV is BGR, which is different from Matplotlib. 
#So, we need to change it to RGB 
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
print(img_rgb.shape)
print(img_rgb[0][0])
```

# 5. Resize, Conversion to Grayscale


```python
#Scale this image to a smaller size
size=(64, 64)
resized_image_feature_vector = cv2.resize(img_rgb, size)
plt.imshow(resized_image_feature_vector)
```

# 7. Face Detection

```python
import cv2, os
import matplotlib.pyplot as plt
import numpy as np
imagePath = 'Image/face.jpg'

#read

face_detection=cv2.CascadeClassifier('Data/opencv/haarcascades/haarcascade_frontalface_alt.xml')
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = face_detection.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(image,(x,y),(x+w, y+h),(255,0,0),3)
plt.imshow(img)
```

# 8.Eye Detection

```python
import cv2, os
import matplotlib.pyplot as plt
import numpy as np
imagePath = 'Image/face.jpg'

eye_detection=cv2.CascadeClassifier('Data/opencv/haarcascades/haarcascade_eye.xml')
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
eyes = eye_detection.detectMultiScale(gray, 1.03, 5)
for (ex,ey,ew,eh) in eyes:
    img = cv2.rectangle(image,(ex,ey),(ex+ew, ey+eh),(0,255,0),2)
plt.imshow(img)
```
